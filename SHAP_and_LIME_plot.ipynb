```python
import matplotlib.pyplot as plt

# Set global font to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

# Data from Table A4: LIME Explanation for Case K-12 (Kerala)
features = [
    "Follicle No. (R)",
    "AMH (ng/mL)",
    "Hair Growth (Y/N)",
    "Skin Darkening (Y/N)",
    "Regular Cycle"
]

weights = [0.43, 0.32, 0.18, 0.15, -0.19]
values = [
    "28 follicles",
    "12.4 ng/mL",
    "Yes",
    "Yes",
    "No"
]

# Colors: Green for positive contribution, Red for negative
colors = ['green' if w > 0 else 'red' for w in weights]

# Reverse for better top-down display
features = features[::-1]
weights = weights[::-1]
colors = colors[::-1]
values = values[::-1]

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(features, weights, color=colors, edgecolor='black')

# Annotate bars with actual feature values
for bar, val in zip(bars, values):
    width = bar.get_width()
    xpos = width + 0.02 if width > 0 else width - 0.25
    plt.text(xpos, bar.get_y() + bar.get_height() / 2, val,
             va='center', ha='center', fontsize=10)

# Labels and title
plt.xlabel("Feature Contribution (Weight)", fontsize=12)
plt.title("LIME Explanation for Case K-12 (Kerala)\nPrediction: 0.92 (PCOS-positive)",
          fontsize=14, weight='bold')

# Add grid
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set global font to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

# Data from Table A4: LIME Explanation for Case K-12 (Kerala)
features = [
    "Follicle No. (R)",
    "AMH (ng/mL)",
    "Hair Growth (Y/N)",
    "Skin Darkening (Y/N)",
    "Regular Cycle"
]

weights = [0.43, 0.32, 0.18, 0.15, -0.19]
values = [
    "28 follicles",
    "12.4 ng/mL",
    "Yes",
    "Yes",
    "No"
]

# For a line plot, keep order as is (no need to reverse)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(features, weights, marker='o', linestyle='-', color='blue')

# Annotate each point with the corresponding feature value
for i, (feat, w, val) in enumerate(zip(features, weights, values)):
    plt.text(i, w + (0.02 if w >= 0 else -0.05), val,
             ha='center', va='bottom' if w >= 0 else 'top', fontsize=10)

# Highlight positive points green, negative red
for i, w in enumerate(weights):
    plt.scatter(i, w, color='green' if w > 0 else 'red', s=100, zorder=5)

plt.xlabel("Features", fontsize=12)
plt.ylabel("Feature Contribution (Weight)", fontsize=12)
plt.title("LIME Explanation for Case K-12 (Kerala)\nPrediction: 0.92 (PCOS-positive)",
          fontsize=14, weight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.ylim(min(weights) - 0.1, max(weights) + 0.1)
plt.tight_layout()
plt.show()
```
